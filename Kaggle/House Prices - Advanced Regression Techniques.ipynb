## **Variance Inflation Factor**



```python
def calculate_vif(df, features):
    vif, tolerance = {}, {}
    
    for feature in features:
        x = [f for f in features if f != feature]
        x, y = df[x], df[feature]

        r2 = LinearRegression().fit(x, y).score(x, y)
        tolerance[feature] = 1 - r2
        vif[feature] = 1 / (tolerance[feature])
    return pd.DataFrame({'VIF':vif, 'Tolerance':tolerance})
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop(['SalePrice'])
calculate_vif(df[numeric_cols], df[numeric_cols].columns).sort_values(by='VIF', ascending=False)
```

- Создаем функцию get_models(), которая возвращает список моделей, где каждая модель определяется кортежем из названия и самой модели.
- Теперь создается новый обучающий набор данных. Т.е. каждая модель из списка get_models() обучается на обучающей выборке, предсказывает значения на тестовой выборке. Такая операция производится для каждой модели, предсказания каждой модели объединяются в новый обучающий набор данных (meta_test_X). Каждый столбец или объект представляет выходные данные одной базовой модели. Каждая строка представляет собой один сэмпл из независимого набора данных.
- После этого нужно обучить смешивающую модель (blender) на meta_test_X и тестовом таргете.
- Для предсказания совершенно новых значений на blender, теперь необходимо предсказать значения для каждой модели из get_models() на валидационной выборке, объединить эти прогнозы в один набор данных и вызвать predict для blender на этом новом наборе данных.

```python
def get_models():
    models = list()
    models.append(('KernelRidge', KernelRidge(**ridge_params)))
    models.append(('GradientBoostingRegressor', GradientBoostingRegressor(**gradient_params)))
    models.append(('XGBRegressor', XGBRegressor(**xgboost_params)))
    models.append(('LGBMRegressor', LGBMRegressor(**lightgbm_params)))
    return models

def fit_ensemble(models, x_train, y_train, x_test, y_test):
    meta_test_X = list()
    for name, model in models:
        model.fit(x_train, y_train)
        yhat = model.predict(x_test)
        yhat = yhat.reshape(len(yhat), 1)
        meta_test_X.append(yhat)

    meta_test_X = np.hstack(meta_test_X)
    display(len(meta_test_X))
    blender = LinearRegression(copy_X=True, fit_intercept=True)
    blender.fit(meta_test_X, y_test)
    return blender

def predict_ensemble(models, blender, x_valid):
    meta_valid_X = list()
    for name, model in models:
        yhat = model.predict(x_valid)
        yhat = yhat.reshape(len(yhat), 1)
        meta_valid_X.append(yhat)
    
    meta_valid_X = np.hstack(meta_valid_X)
    return blender.predict(meta_valid_X)
```
