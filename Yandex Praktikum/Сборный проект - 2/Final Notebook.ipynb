# Описание данных.
## 1. Технологический процесс.
- *Rougher feed* — исходное сырье
- *Rougher additions* (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- *Xanthate* — ксантогенат (промотер, или активатор флотации);
- *Sulphate* — сульфат (на данном производстве сульфид натрия);
- *Depressant* — депрессант (силикат натрия).
- *Rougher process* (англ. «грубый процесс») — флотация
- *Rougher tails* — отвальные хвосты
- *Float banks* — флотационная установка
- *Cleaner process* — очистка
- *Rougher Au* — черновой концентрат золота
- *Final Au* — финальный концентрат золота

## 2. Параметры этапов.
- *air amount* — объём воздуха
- *fluid levels* — уровень жидкости
- *feed size* — размер гранул сырья
- *feed rate* — скорость подачи

## 1.3. Проверка правильности расчета эффективности обогащения

```python
df_train.dropna(subset = ['rougher.output.recovery'], inplace = True)

recovery_C = df_train['rougher.output.concentrate_au'] 
recovery_F = df_train['rougher.input.feed_au']
recovery_T = df_train['rougher.output.tail_au']

recovery_new = ((recovery_C * (recovery_F - recovery_T)) / (recovery_F * (recovery_C - recovery_T))) * 100
recovery_old = df_train['rougher.output.recovery']
```

```python
ax = plt.gca()
xlims = [(recovery_old.min(), recovery_old.max()), (recovery_new.min(), recovery_new.max())]
title = 'Гистограмма сравнения показателей эффективности обогащения, %'
xlim1 = xlims[0]
xlim2 = xlims[1]
recovery_old.plot(kind='hist', bins=range(300), figsize=(12,6), ec='black', ax=ax, color='#FA7D79', 
                             title=title, xlim=xlim1, alpha = 0.5)
recovery_new.plot(kind='hist', bins=range(300), figsize=(12,6), ec='black', ax=ax, color='#F6FA91', 
                             title=title, xlim=xlim2, alpha = 0.5)

plt.xlabel('Эффективность обогащения', fontsize=14)
plt.ylabel('Частота показателя', fontsize=14)
plt.legend(['Исходный показатель эффективности обогащения',
            'Новый показатель эффективности обогащения'],
           prop={'size':10})
plt.show()
print('MAE:', mean_absolute_error(recovery_old, recovery_new))
```

Исходя из гистограммы наблюдаем что эффективность обогащения была рассчитана максимально верно, средняя абсолютная ошибка стремится к минимуму. Теперь можем переходить к удалению лишних столбцов из обучающей выборки, и добавить столбцы rougher.output.recovery и final.output.recovery в тестовую выборку.

# 2. Анализ данных.

## 2.1. Распределение концентрации металлов на различных этапах очистки.

```python
stages = [('rougher.input.feed_', 'Исходное сырье'), ('rougher.output.concentrate_', 'Флотация'), 
          ('primary_cleaner.output.concentrate_', 'Первичный этап очистки'), ('final.output.concentrate_', 'Вторичный этап очистки')]

metals = [('ag', 'серебра'), ('pb', 'свинца'), ('au', 'золота')]
def distribution_plotting(stages, metals, df):
    for metal_name in metals:
        plt.figure(figsize=(16,6))
        for stage in stages:
            column_name = stage[0] + metal_name[0]
            cleared_df = df_full[column_name]
            cleared_df = cleared_df
            ax = sns.distplot(cleared_df, label=stage[1])
            plt.legend(prop={'size':10}, loc='upper right')
            ax.set(xlabel='Распределение концeнтрации ' + metal_name[1], 
                   title='Изменение концентрации ' + metal_name[1] + ' на всех производственных этапах')
        plt.show()
```

```python
distribution_plotting(stages, metals, df_full)
```

Исходя из полученных графиков, мы можем наблюдать следующие тенденции:
- **Концентрация серебра сильно варьируется между этапами производственного процесса**: на этапе флотации достигается своего максимума, однако с переходом на очистку концентрация серебра начинает снижаться с каждой последующей очисткой.

- **Концентрация свинца плавно возрастает** по мере продвижения в производственном процессе: однако существенных различий между концентрацией на первичном этапе очистки и концентрации на вторичном этапе очистки не наблюдается.

- **Концентрация золота существенно возрастает с каждым этапом** производственного процесса в среднем на 10%.

- Также присутствуют выбросы на распределениях всех веществ в районе 0-1%. Для более качественного обучения модели, необходимо избавиться от этих значений.

```python
print('Размер исходного датасета до обработки выбросов -', df_full.shape)
for metal_name in metals:
    for stage in stages:
        column_name = stage[0] + metal_name[0]
        df_full = df_full.drop(index = df_full[df_full[column_name] < 1.75].index, axis = 0)
print('Размер исходного датасета после обработки выбросов -', df_full.shape)
```

Выбросы удалены, теперь взглянем на распределение размеров гранул сырья на обучающей и тестовой выборке для оценки

# 3. Построение модели.

## 3.1. Создание функции для вычисления итоговой sMAPE.

Перед непосредственно построением моделей, подготовим функцию для вычисления нашей ключевой метрики - sMAPE.

**sMAPE** - Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение». В отличии от метрики MAE выражается в относительных величинах, симметричное - так как одинаково учитывает масштаб и целевого признака, и предсказания.

```python
def smape(y_true, y_pred):
    result = np.sum(abs(y_true - y_pred) / ((abs(y_true) + abs(y_pred)) / 2)) * 100/len(y_true)
    return result

def smape_final(rougher, final):
    return  0.25*rougher + 0.75*final
```

## 3.2. Обучение различных моделей и оценка их качества с помощью кросс-валидации.

В качестве моделей будем использовать линейную регрессию, дерево решений, и случайный лес. Кроссвалидацию будем проводить внутри GridSearchCV.

Предварительно разделим наши обучающую и тестовую выборки еще на две: для предсказания эффективности обогащения чернового концентрата (rougher.output.recovery) и эффективности обогащения финального концентрата (final.output.recovery) и после этого разделим их по признакам и целевому признаку. Так же мы избавимся от столбца 'date' за ненадобностью и масштабируем признаки.

```python
rougher_columns = []
for i in list(df_train.columns):
    if 'rougher' in i:
        rougher_columns.append(i)
```

```python
df_train = df_train.drop('date', axis = 1)
df_test = df_test.drop('date', axis = 1)

df_train_rougher = df_train[rougher_columns]
df_test_rougher = df_test[rougher_columns]

df_train_final = df_train.drop(rougher_columns, axis = 1)
df_test_final = df_test.drop(rougher_columns, axis = 1)
```

Выборки разделены по двум целевым признакам, теперь еще раз разделим их на признаки и целевые признаки.

```python
x_train_rougher = df_train_rougher.drop('rougher.output.recovery', axis = 1)
x_test_rougher = df_test_rougher.drop('rougher.output.recovery', axis = 1)
y_train_rougher = df_train_rougher['rougher.output.recovery']
y_test_rougher = df_test_rougher['rougher.output.recovery']

x_train_final = df_train_final.drop('final.output.recovery', axis = 1)
x_test_final = df_test_final.drop('final.output.recovery', axis = 1)
y_train_final = df_train_final['final.output.recovery']
y_test_final = df_test_final['final.output.recovery']

xy_rougher_list = [(x_train_rougher, y_train_rougher), (x_test_rougher, y_test_rougher)]
xy_final_list = [(x_train_final, y_train_rougher), (x_test_final, y_test_final)]
```

Теперь масштабируем признаки с помощью StandardScaler

```python
scaler = StandardScaler()
scaler.fit(x_train_rougher)
x_train_rougher = scaler.transform(x_train_rougher)
x_test_rougher = scaler.transform(x_test_rougher)

scaler = StandardScaler()
scaler.fit(x_train_final)
x_train_final = scaler.transform(x_train_final)
x_test_final = scaler.transform(x_test_final)

xy_rougher_list = [(x_train_rougher, y_train_rougher), (x_test_rougher, y_test_rougher)]
xy_final_list = [(x_train_final, y_train_final), (x_test_final, y_test_final)]
```

Переходим непосредственно к обучению и тестированию наших моделей. Для этого напишем функцию которая на входе будет получать список с тренировочными и тестовыми выборками , наименование используемой модели, и словарь с гиперпараметрами для реализуемой модели. Внутри функция будет выполнять следующие шаги:

1. Проведет автоподбор гиперпараметров с помощью GridSearchCV на тренировочных выборках с двойной кросс-валидаций,
2. Обучится и предскажет нам результаты на тестовой выборке.
3. Рассчитает sMAPE и итоговый sMAPE на тренировочной и тестовой выборке.
4. На основе sMAPE функция также проведет выбор наилучшего результата, и выведет на результаты sMAPE на обучающей и тестовой выборках на этапах флотации и получения финальных характеристик, наименование модели, и затрачиваемое время на выполнение кода.

```python
smape_results = pd.DataFrame({'Модель' : [], 'Train sMAPE score' : [], 'Test sMAPE score' : [], 'Purpose' : []})
def model_testing(model_name, parametrs):
    
    global smape_results
    
    random_state = 12345
    
    cv = KFold(n_splits=5, shuffle=False, random_state = 12345)
    smape_scorer = make_scorer(smape, greater_is_better = False)
    
    if model_name == DecisionTreeRegressor:
        output_model_name = 'Дерево решений'
    elif model_name == LinearRegression:
        output_model_name = 'Линейная регрессия'
    elif model_name == RandomForestRegressor:
        output_model_name = 'Случайный лес'
    
    train_smape_rough = 0
    test_smape_rough = 0
    train_smape_final = 0
    test_smape_final = 0
    
    if output_model_name == 'Линейная регрессия':
        model = model_name(n_jobs = -1)
    elif output_model_name == 'Случайный лес':
        model = model_name(n_jobs = -1, random_state = random_state)
    else:
        model = model_name(random_state = random_state)

    grid = GridSearchCV(model, parametrs, cv = cv, scoring = smape_scorer, n_jobs = -1)

    for i in range(0, 2):
        if i == 0:
            
            if output_model_name == 'Линейная регрессия':
                grid.fit(xy_rougher_list[0][0], xy_rougher_list[0][1])
                test_model = model_name(**grid.best_params_)
            else:
                grid.fit(xy_rougher_list[0][0], xy_rougher_list[0][1])
                test_model = model_name(**grid.best_params_, random_state = random_state)
            
            test_model.fit(xy_rougher_list[0][0], xy_rougher_list[0][1])
            model_train_predictions = test_model.predict(xy_rougher_list[0][0])
            model_test_predictions = test_model.predict(xy_rougher_list[1][0])
            train_smape_rough = smape(xy_rougher_list[0][1], model_train_predictions)
            test_smape_rough = smape(xy_rougher_list[1][1], model_test_predictions)
            smape_results = smape_results.append({'Модель' : output_model_name, 
                                  'Train sMAPE score' : train_smape_rough,
                                  'Test sMAPE score' : test_smape_rough,
                                  'Purpose' : 'Rougher'}, ignore_index = True)
            print('\033[1m' +'Модель -', output_model_name + '\033[0m')
            print('')
            print('Производственный этап : ' + '\033[1m' + 'Флотация' + '\033[0m')
            print('Train sMAPE score -', train_smape_rough)
            print('Test sMAPE score -', test_smape_rough)
        elif i == 1:  
            
            if output_model_name == 'Линейная регрессия':
                test_model = model_name(**grid.best_params_)
                grid.fit(xy_final_list[0][0], xy_final_list[0][1]) 
            else:
                test_model = model_name(**grid.best_params_, random_state = random_state)
                grid.fit(xy_final_list[0][0], xy_final_list[0][1]) 
            
            test_model.fit(xy_final_list[0][0], xy_final_list[0][1])
            model_train_predictions = test_model.predict(xy_final_list[0][0])
            model_test_predictions = test_model.predict(xy_final_list[1][0])
            train_smape_final = smape(xy_final_list[0][1], model_train_predictions)
            test_smape_final = smape(xy_final_list[1][1], model_test_predictions)
            smape_results = smape_results.append({'Модель' : output_model_name, 
                                  'Train sMAPE score' : train_smape_final,
                                  'Test sMAPE score' : test_smape_final,
                                  'Purpose' : 'Final'}, ignore_index = True)
            print('')
            print('Производственный этап : ' + '\033[1m' + 'Финальные характеристики' + '\033[0m')
            print('Train sMAPE score -', train_smape_final)
            print('Test sMAPE score -', test_smape_final)
            smape_results = smape_results.append({'Модель' : output_model_name,
                                                 'Train sMAPE score' : smape_final(train_smape_rough, train_smape_final),
                                                 'Test sMAPE score' : smape_final(test_smape_rough, test_smape_final),
                                                 'Purpose' : 'Итоговый расчет sMAPE'}, ignore_index = True)
```

Функция готова к применению, используем её на наших выбранных моделях с определённым набором гиперпараметров.

```python
%%time
parametrs = {'max_depth': range (1, 15, 2),
             'min_samples_leaf': range (2,6),
             'min_samples_split': range (2,4)}
model_testing(DecisionTreeRegressor, parametrs)
```

```python
%%time
parametrs = {'max_depth': range (3, 15, 5),
             'min_samples_leaf': range (2, 5),
             'min_samples_split': range (2, 5),
             'n_estimators': range (2, 10, 2)}
model_testing(RandomForestRegressor, parametrs)
```

```python
%%time
parametrs = {'fit_intercept':('True', 'False'), 
             'normalize':('True', 'False'), 
             'copy_X':('True', 'False')}

model_testing(LinearRegression, parametrs)
```
